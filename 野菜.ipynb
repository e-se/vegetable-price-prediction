```python
import numpy as np
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
```

```python
# googleドライブマウント
from google.colab import drive
drive.mount('/content/drive')

# データの読み込み
train=pd.read_csv("/content/drive/MyDrive/competition/vegetable/train_data.csv")
test_sub=pd.read_csv("/content/drive/MyDrive/competition/vegetable/submission.csv")
weather= pd.read_csv("/content/drive/MyDrive/competition/vegetable/weather.csv")
```

```python
test_sub
```

```python
# 品目リストを作成
item_name_list=test_sub['id'].values.tolist()
```

```python
# 予測をする関数を作成する
def price_predict(item_name: str):
  # 学習データの読み込み
  train_df=train[["id",item_name]]
  train_df.columns=['ds', 'y']
  model = Prophet()

  # モデルをデータにフィットさせます
  model.fit(train_df)

  future = {'ds': ['2019-12-01'] }
  future = pd.DataFrame(data=future)

  # 未来の日付に対する予測を行います
  forecast = model.predict(future)

  return forecast["yhat"].to_string(index=False)
```

```python
sub_data = pd.DataFrame(columns=['id','y'])

# 予測とsubmissionデータ作成
index_number=0
for item_name in item_name_list:
  # 値の予測
  item_price = price_predict(item_name)
  # submissionデータ作成
  sub_data.loc[index_number] =[item_name,item_price]
  index_number +=1
```

```python
# ファイルに出力
sub_data.to_csv("/content/drive/MyDrive/competition/vegetable/sub_20230817.csv", index=False)
```
