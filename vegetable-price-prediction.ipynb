```python
# 
import numpy as np
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# 
# (ds)(y)2
# 
data = {
    'ds': pd.date_range(start='2022-01-01', end='2023-01-01', freq='D'),
    'y': (20 * (1 + 0.05 * np.random.randn(366))).cumsum()
}
df = pd.DataFrame(data)

# Prophet
model = Prophet()

# 
model.fit(df)

# 
# 1
future = model.make_future_dataframe(periods=365)

# 
forecast = model.predict(future)

# 
fig = model.plot(forecast)
plt.show()
```
